```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
datadir = '../../../data/'
infodict = {
    'gc':
    {
        'longname':'globcolour',
        'filename_wmean-latlon':'globcolour-cmems.day-tomonthly.wmean-latlon.JANtoDEC.1998-2023.ForBuchovecky.nc',
        'conversion':1
    },
    'esh':
    {
        'longname':'esm2m-historical',
        'filename':'esm2m-historical.monthly.wmean-latlon.JANtoDEC.1991-2005.ForBuchovecky.nc',
        'conversion':1035*1000
    },
    'esc':
    {
        'longname':'esm2m-control',
        'filename':'esm2m-control.monthly.wmean-latlon.JANtoDEC.0001-0300.ForBuchovecky.nc',
        'conversion':1035*1000
    } 
}
```

```python
dd = {}
for key in infodict.keys():
    path = datadir+infodict[key]['filename']
    dd[key] = xr.open_dataset(path)
```

```python
fig,ax = plt.subplots(figsize=(8,4))
for key in infodict.keys():
    v = dd[key]['PanAntarctic']*infodict[key]['conversion']
    ax.plot(v)
```
